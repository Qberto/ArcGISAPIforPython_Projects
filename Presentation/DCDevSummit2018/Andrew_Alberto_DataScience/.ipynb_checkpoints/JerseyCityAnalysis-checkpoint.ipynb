```python
from arcgis.features import SpatialDataFrame
from arcgis.features._data.geodataset import spatial_join
import pandas as pd
import pysal as ps
import numpy as np
```

```python
csv_crime = pd.read_csv(r"./data/jersey_city/jcpd-all.csv", 
                        index_col=None, 
                        infer_datetime_format=True)
```

```python
csv_crime.head(2)
```

```python
### Get the Drunk/Minor offenses Data
#q = csv_crime['Call Code Description'].str.lower().str.contains('drunk')
df = pd.DataFrame(csv_crime['Call Code Description'].str.split(';').tolist(), 
                  columns=['Description', 'TimeElapse', 'InjuryDescription'])
csv_crime['LABELS'] = df['Description']
csv_crime['LABELS'].head(2)
```

```python
vals = ['INTOXICATED PERSON', 'INTOXICATED DRIVER', 'LIQUOR POSSESSION BY MINOR']
q = csv_crime['LABELS'].isin(vals)
alcohal_df = csv_crime[q].copy()
alcohal_df.reset_index(drop=True, inplace=True)
alcohal_df.head(2)
```

## Convert to Spatial DataFrame



```python
from arcgis.geometry import Point
geoms = [Point({'x': x, 'y': y, 'spatialReference' : {'wkid' : 4326}}) \
         for x,y in alcohal_df[['LONGITUDE','LATITUDE']].to_records(index=False)]
sdf = SpatialDataFrame(data=alcohal_df, geometry=geoms)
```

```python
sdf.head(1)
```

```python
pd.to_datetime(sdf['Time Received']
              ).dt.hour.value_counts().sort_index(
).plot(
    'bar', 
    title='Incidents by Hour')
```

```python
res = arcgis.geocoding.batch_geocode(addresses=booze_df['Premise Address'].tolist(),
                             as_featureset=False)
```

```python
res
```

```python
sr = {'wkid' : 4326}
locations  = pd.DataFrame(res)

geoms = []
for g in locations['location'].tolist():
    g['spatialReference'] = sr
    geoms.append(Point(g))
bars_df = SpatialDataFrame(data=locations, geometry=geoms)
```

```python
attribute_joined_bars = bars_df.reset_index(inplace=False, drop=True).join(booze_df.reset_index(inplace=False, drop=True))
```

```python
census_df = SpatialDataFrame.from_featureclass(
    r"./data/jersey_city/JCBlockGroups1984.shp")
#census_df = SpatialDataFrame.from_layer(layer=item.layers[0])
#census_df.geometry = census_df.project_as(4326)
census_df.head(2)
```

```python

joined = spatial_join(df1=bars_df, df2=census_df)
len(joined)
```

```python
counts = joined['GEOID'].value_counts().to_frame().reset_index(inplace=False, drop=False)

counts['count'] = counts['GEOID']
del counts['GEOID']
counts['GEOID'] = counts['index']
del counts['index']
```

```python
counts.columns
```

```python
merged = census_df.merge(counts, how="outer")
merged.head()
```

```python
q = merged['count'].isnull()
merged['count'].fillna(0, inplace=True)
```

```python
## calculate the bar per square mile
merged['BarPerSqMi'] = merged['count'] / merged.geometry.get_area("GEODESIC", units="SQUAREMILES") 
```

```python
q = merged['BarPerSqMi'] == np.inf
merged.loc[q, 'BarPerSqMi'] = 0
```

```python
m3 = GIS().map()
m3
```

```python
merged.plot(kind='map', map_widget=m3)
```
