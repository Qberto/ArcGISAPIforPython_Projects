```python
from arcgis.features import SpatialDataFrame
from arcgis.features._data.geodataset import spatial_join
import pandas as pd
import pysal as ps
import numpy as np
```

```python
csv_crime = pd.read_csv(r"./data/jersey_city/jcpd-all.csv", 
                        index_col=None, 
                        infer_datetime_format=True)
df = pd.DataFrame(csv_crime['Call Code Description'].str.split(';').tolist(), 
                  columns=['Description', 'TimeElapse', 'InjuryDescription'])
csv_crime['LABELS'] = df['Description']
csv_crime['LABELS'].head(2)
```

```python
vals = ['INTOXICATED PERSON', 'INTOXICATED DRIVER', 'LIQUOR POSSESSION BY MINOR']
q = csv_crime['LABELS'].isin(vals)
alcohal_df = csv_crime[q].copy()
alcohal_df.reset_index(drop=True, inplace=True)
alcohal_df.head()
```

```python
from arcgis.geometry import Point
geoms = [Point({'x': x, 'y': y, 'spatialReference' : {'wkid' : 4326}}) \
         for x,y in alcohal_df[['LONGITUDE','LATITUDE']].to_records(index=False)]
offenses_df = SpatialDataFrame(data=alcohal_df, geometry=geoms)
offenses_df.head()
```

```python
bars_df = SpatialDataFrame.from_featureclass('c:\\temp\\scratch.gdb\\jersey_city_bars')
census_df = SpatialDataFrame.from_featureclass(r"./data/jersey_city/JCBlockGroups1984.shp")
demographics_df = pd.read_csv(r"Data/jersey_city/ACS/demographics.csv")
demographics_df.index = demographics_df['GEOID'].astype(str)
```

## Spatially Join Bars and Census Data

- get the counts


```python
joined = spatial_join(df1=bars_df, df2=census_df)
counts = joined['GEOID'].value_counts().to_frame().reset_index(inplace=False, drop=False)
counts['count'] = counts['GEOID']
del counts['GEOID']
counts['GEOID'] = counts['index']
del counts['index']
merged = census_df.merge(counts, how="outer")
```

```python
census_df_with_counts = merged
census_df_with_counts['count'].fillna(0, inplace=True)
census_df_with_counts[['GEOID', 'count']].head()
```

```python
census_df_with_counts['BarPerSqMi'] = census_df_with_counts['count'] / census_df_with_counts.geometry.get_area("GEODESIC", 
                                                                                                               units="SQUAREMILES") 
```

```python
q = census_df_with_counts['BarPerSqMi'] == np.inf
census_df_with_counts.loc[q, 'BarPerSqMi'] = 0
census_df_with_counts[['GEOID', 'count', 'BarPerSqMi']].head()
```

```python
census_df_with_counts.index = census_df_with_counts.GEOID.astype(str)
census_df_with_counts
```
