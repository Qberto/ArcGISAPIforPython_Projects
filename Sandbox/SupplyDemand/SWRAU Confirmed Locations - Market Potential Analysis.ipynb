### Create Scoring Indexes

```python
commercial_tech_scoring_index = {
    "POP18UP_CY": 0.5,
    "CIVLBFR_CY": 0.1,
    "INDINFO_CY": 0.2,
    "INDTECH_CY": 0.2
}

industrial_scoring_index = {
    "CIVLBFR_CY": 0.10,
    "INDMIN_CY": 0.30,
    "INDCONS_CY": 0.30,
    "INDMANU_CY": 0.30,
}

agriculture_scoring_index = {
    "CIVLBFR_CY": 0.10,
    "INDAGRI_CY": 0.75
}

employment_scoring_index = {
    "UNEMP_CY": 0.75,
    "HINC0_CY": 0.25
}
```

```python
def calc_commercial_tech_scoring_index(pop18up_cy, pop18up_cy_ix, civilbfr_cy,civilbfr_cy_ix,indinfo_cy, indinfo_cy_ix, indtech_cy, indtech_cy_ix):
    return pop18up_cy * pop18up_cy_ix + civilbfr_cy * civilbfr_cy_ix + indinfo_cy * indinfo_cy_ix + indtech_cy *indtech_cy_ix
```

```python
def calc_industrial_scoring_index(CIVLBFR_CY, CIVLBFR_CY_ix, INDMIN_CY, INDMIN_CY_ix, INDCONS_CY, INDCONS_CY_ix, INDMANU_CY, INDMANU_CY_ix):
    return CIVLBFR_CY * CIVLBFR_CY_ix + INDMIN_CY * INDMIN_CY_ix + INDCONS_CY * INDCONS_CY_ix + INDMANU_CY *INDMANU_CY_ix
```

```python
def calc_agricultural_scoring_index(CIVLBFR_CY, CIVLBFR_CY_ix, INDAGRI_CY, INDAGRI_CY_ix):
    return CIVLBFR_CY * CIVLBFR_CY_ix + INDAGRI_CY * INDAGRI_CY_ix
```

```python
def calc_employment_scoring_index(UNEMP_CY, UNEMP_CY_ix, HINC0_CY, HINC0_CY_ix):
    return UNEMP_CY * UNEMP_CY_ix + HINC0_CY * HINC0_CY_ix
```

```python
psa_nodes_df['commercial_tech_score'] = psa_nodes_df.apply(lambda row: calc_commercial_tech_scoring_index(row['POP18UP_CY_12_zscore'], commercial_tech_scoring_index['POP18UP_CY'], row['CIVLBFR_CY_zscore'], commercial_tech_scoring_index['CIVLBFR_CY'], row['INDINFO_CY_zscore'], commercial_tech_scoring_index['INDINFO_CY'], row['INDTECH_CY_zscore'], commercial_tech_scoring_index['INDTECH_CY']), axis=1)
```

```python
psa_nodes_df['industrial_score'] = psa_nodes_df.apply(lambda row: calc_industrial_scoring_index(row['CIVLBFR_CY_zscore'], industrial_scoring_index['CIVLBFR_CY'], row['INDMIN_CY_zscore'], industrial_scoring_index['INDMIN_CY'], row['INDCONS_CY_zscore'], industrial_scoring_index['INDCONS_CY'], row['INDMANU_CY_zscore'], industrial_scoring_index['INDMANU_CY']), axis=1)
```

```python
psa_nodes_df['agricultural_score'] = psa_nodes_df.apply(lambda row: calc_agricultural_scoring_index(row['CIVLBFR_CY_zscore'], agriculture_scoring_index['CIVLBFR_CY'], row['INDAGRI_CY_zscore'], agriculture_scoring_index['INDAGRI_CY']), axis=1)
```

```python
psa_nodes_df['employment_score'] = psa_nodes_df.apply(lambda row: calc_employment_scoring_index(row['UNEMP_CY_zscore'], employment_scoring_index['UNEMP_CY'], row['HINC0_CY_zscore'], employment_scoring_index['HINC0_CY']), axis=1)
```
